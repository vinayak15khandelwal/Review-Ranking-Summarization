```python
import pandas as pd
import json


# FUNCTION TO LOAD LIMITED NUMBER OF REVIEWS
def load_data(path, limit):
    
    data = []
    
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        
        for i, line in enumerate(f):
            
            if i >= limit:
                break
                
            data.append(json.loads(line))
    
    
    df = pd.DataFrame(data)
    
    return df


# CHANGE THIS VALUE WHEN YOU WANT DIFFERENT DATA SIZE
DATA_LIMIT = 5000


# LOAD DATASET
df = load_data("../data/reviews_Electronics_5.json", DATA_LIMIT)

print("Rows loaded:", len(df))


# DISPLAY SAMPLE
df.tail()
```

```python
import os

# Ensure data folder exists
os.makedirs("../data", exist_ok=True)

# Save dataset
df.to_csv("../data/raw_reviews.csv", index=False)

print("Raw limited dataset saved:", len(df))
```
